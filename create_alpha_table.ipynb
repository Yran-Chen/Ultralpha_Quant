```python
param_dir = ".." + os.sep + "eval_opt_res" + os.sep + "crosssec"
```

```python
param_df = pd.DataFrame()

i = 0 
for stra in os.listdir(param_dir):
    stra_path = os.path.join(param_dir, stra)
    for res_dir in os.listdir(stra_path):
        res_path = os.path.join(stra_path, res_dir)
        file_path = os.path.join(res_path, "param.txt")
        res_path = os.path.join(res_path, "res.csv" )
        
        print("current param res dir: ", res_dir)
       
        # parse the parameters
        config = configparser.ConfigParser()
        config.read(file_path)
        params = config["params"]
        dic = {}
        for item, value in params.items():
            if item == "start_date" or item == "end_date":
                value = value.split(" ")[0]
                dic[item] = datetime.datetime.strptime(value, "%Y-%m-%d")
            elif item in ["indicator_cal_method", "freq_unit", "res_dir", "index_col", "fut_pool_gen_method"]:
                dic[item] = value
                pass
            else:
                dic[item] = eval(params[item])
        
        # parse the result 
        res = pd.read_csv(res_path)
        res = res.set_index(["indi"])
        for val in ["max_drawdown", "sharpe", "returns", "ir", "std"]:
            dic[val] = res.loc[val, "value"]
#         print(res.loc["max_drawdown", "value"], res.loc["sharpe", "value"])
        param_df = param_df.append(pd.DataFrame([dic]), ignore_index=True)
        print(dic)
#         print(i)
#         print(" ")
#         print(" ")
        
#         strategy_param["crosssec_" + str(i)] = dic.copy()
        i += 1
        print(i)
```

# create the alpha table

```python
cal_indi_param_df = param_df[alpha_elements]
cal_indi_param_df = cal_indi_param_df.rename(columns = {"indicator_cal_method": "method"})
cal_indi_param_df["alpha_id"] = cal_indi_param_df["method"]
cal_indi_param_df
```

```python
alpha_dic = {"hedging_pressure": "hp",
             "inflation": "in",
             "momentum": "mom",
             "open_interest": "oi",
             "term_structure": "ts"}
```

```python
cal_indi_param_df["alpha_id"] = cal_indi_param_df["alpha_id"].map(alpha_dic) 
```

```python
cal_indi_param_df = cal_indi_param_df.drop_duplicates(subset = cal_indi_param_df.columns)
```

```python
def assign(sub_df):
    sub_df["id"] = np.arange(sub_df.shape[0])
    return sub_df 

cal_indi_param_df = cal_indi_param_df.groupby(["method"]).apply(assign)
```

```python
cal_indi_param_df["alpha_id"] = cal_indi_param_df.apply(lambda r:r["alpha_id"] +   "-" + str(r["id"]), axis=1 )
cal_indi_param_df = cal_indi_param_df.drop(columns = ["id"])
```

```python
for col in ["num_days", "num_durations", "freq"]:
    cal_indi_param_df[col] = cal_indi_param_df[col].map(lambda x: int(x) if not np.isnan(x) else x)
# cal_indi_param_df["freq"] = cal_indi_param_df["freq"].map(lambda x: np.asscalar(int(x)) if not np.isnan(x) else x)
```

```python
indi_list = []
for i in cal_indi_param_df.index:
    param = cal_indi_param_df.loc[i]
    param_dic = {index: param[index] for index in param.index}
    alpha_id = cal_indi_param_df.loc[i, "alpha_id"]
    param_dic["start_date"] = start_date
    param_dic["end_date"] = end_date
    param_dic["future_pool"] = None
    param_dic["freq"] = param_dic["freq"].item()
    print(param_dic)
#     print(type(param_dic["freq"]))
    indi = indi_generator.cal_indicator(param_dic)
    indi = indi.rename(columns = {"indicator": alpha_id})
    print(" " * 7)
#     print(indi)
    indi_list.append(indi)
```
