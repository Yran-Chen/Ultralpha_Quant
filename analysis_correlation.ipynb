## compare opt result

```python
opt_res_dir = ".." + os.sep + "opt_res"
print(os.listdir(opt_res_dir))
```

```python
config = configparser.ConfigParser()
result = pd.DataFrame()
for file in os.listdir(opt_res_dir):
    res_path = os.path.join(opt_res_dir, file)
    indi_path = os.path.join(res_path, "indi_df.csv")
    param_path = os.path.join(res_path, "param.txt")
    config.read(param_path)
    freq = config["params"]["freq"]
    top = config["params"]["top"]
    indi_met = config["params"]["indicator_cal_method"]
    dic = {"freq": [freq],"top": [top], "indi_met": [indi_met], "indi_file": [indi_path], "param_file": [param_path]}
    result = result.append(pd.DataFrame(dic), ignore_index=True)
```

```python
ascending_dic = {"hedge_pressure": False,
                "inflation": False,
                "open_interest": False}
def assign(g):
    # todo : ascending flexible ?
#     g = g.sort_values( by = ['indicator'], ascending = False)
    g["rank"] = g["indicator"].rank(ascending = False)
    g = g.drop(columns = ["indicator"], axis =1)
    return  g

def cal_spearmanr(g):
#     print(g[:2])
#     print(scipy.stats.spearmanr(g["rank_x"], g["rank_y"]))
    corr = scipy.stats.spearmanr(g["rank_x"], g["rank_y"])
    g["correlation"] = corr[0]
    return g

corr_df = pd.DataFrame()

for  (freq, top), group in result.groupby(["freq", "top"]):
    if group.shape[0] <= 1:
        continue
    num = 0
    indi_list = []
    for i in group.index:
#         print(" current res index {} ".format(i))
        indi_met = result.loc[i, "indi_met"]
        indi_file =  result.loc[i, "indi_file"]
        indi = pd.read_csv(indi_file)
        if "Unnamed: 0" in indi.columns:
            indi = indi.drop(columns = ["Unnamed: 0"], axis = 1)
        if "Unnamed: 1" in indi.columns:
            indi = indi.drop(columns = ["Unnamed: 1"], axis = 1)
        indi = indi.groupby(["date"]).apply(assign)
        indi_list.append(indi)
#     print(group.shape[0])
    alist = [_ for _ in range(group.shape[0])]
    for pair in itertools.combinations(alist, r=2):
        i, j = pair[0], pair[1]
        indi_0, indi_1 = indi_list[i], indi_list[j]
        indi_col = pd.merge(indi_0, indi_1, on=["date", "instrument"])
        indi_col = indi_col.groupby("date").apply(cal_spearmanr)
        indi_col = indi_col.drop_duplicates(subset = "date", keep = "first")
#         print(group.index[i], group.index[j])
#         print(indi_col["correlation"].mean())
        index_0, index_1 = group.index[i], group.index[j]
        met_0, met_1 = result.loc[index_0, "indi_met"], result.loc[index_1, "indi_met"]
        res_0, res_1 = result.loc[index_0, "param_file"].split("\\")[-2], result.loc[index_1, "param_file"].split("\\")[-2]
        dic = {"met_0": [met_0], "met_1": [met_1], "res_0": [res_0], "res_1": [res_1], "corr":[indi_col["correlation"].mean()]}
        corr_df = corr_df.append(pd.DataFrame(dic), ignore_index=True)
```

```python
opt_res_dir = ".." + os.sep + "eval_opt_res" + os.sep + "crosssec"
print(os.listdir(opt_res_dir))
```

```python
i = 0
res_df = pd.DataFrame()

for stra in os.listdir(opt_res_dir):
    stra_path = os.path.join(opt_res_dir, stra)
    
    for res in os.listdir(stra_path):
        res_path = os.path.join(stra_path, res)
        pnl_file = os.path.join(res_path, "daily_pnl.csv")
        res_path = os.path.join(res_path, "res.csv" )
        dic = {}
        
        # read   pnl
        pnl = pd.read_csv(pnl_file)
        pnl = pnl.rename(columns = {"pnl": stra + "_" + res[3:]})
        if i == 0:
            merge_pnl = pnl
        else:
            merge_pnl = pd.merge(merge_pnl, pnl, on=["date"])
            
        # read result
        res_stra_df = pd.read_csv(res_path)
        res_stra_df = res_stra_df.set_index(["indi"])
        for val in ["max_drawdown", "sharpe", "returns", "ir", "std"]:
            dic[val] = res_stra_df.loc[val, "value"]
        dic["alpha"] = stra + "_" + res[3:]
        dic["dir"] =  res
        res_df = res_df.append(pd.DataFrame([dic]), ignore_index=True)

        i += 1
```
